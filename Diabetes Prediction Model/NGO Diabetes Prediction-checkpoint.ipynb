## PROBLEM IDENTIFICATION

We identify an NGO focused on Women and Children affairs.The NGO has recorded high rate of diabetes infection among women.
The object is to diagnostically predict wether a woman has diabetes or not based on data collected. 


## DATA

The data on this problem was sourced from National Insitute of Diabetes and Digestive and Kidney diesease, with the aim of predicted Diabetes status of a woman.

Pregnancy = Number of times pregnant
Glucose = Plasma glucose concentration a 2 hours in an oral glucose tolerance test
Blood Pressure = Diastolic blood pressure (mm Hg)
Skin thickness = Triceps skin fold thickness (mm)
Insulin = 2-Hour serum insulin (mu U/ml)
Body Mass Index (BMI) = Body mass index (weight in kg/(height in m)^2)
Diabetes Pedigree Function = Diabetes pedigree function
Age = Age (years)
Outcome = Class variable (0 or 1) 268 of 768 are 1, the others are 0 1 = Presence of Diabetes and 0 = Abscence of Diabetes

## Modeling

```python
#Importing models
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

```python
X = Data.drop('Outcome',axis=1)
robust_scaler = RobustScaler()
X = robust_scaler.fit_transform(X)
y = Data['Outcome']
X_train,X_test,y_train,y_test =train_test_split(X,y,test_size=0.4,random_state=43,stratify=y)
```

```python
Data.isnull().sum()
```

#### Logistic Regression

```python
#Training Logistic Regression model
Lr = LogisticRegression()
Lr.fit(X_train,y_train)
y_train = Lr.predict(X_train)
```

```python
#Testing the Logistic regression model
predictions = Lr.predict(X_test)
Lr.score(X_test,y_test)
```

```python
#Checking Model Accuracy and Precision score
train_accuracy = accuracy_score(y_test,predictions)
train_precision = precision_score(y_test,predictions)
print("The accuracy is %0.2f" % train_accuracy + "%")
print("The precision is %0.2f" % train_precision + "%")
```

```python
#Confusion matrix
cnf_matrix = confusion_matrix(y_test,predictions)
pd.crosstab(y_test, predictions, rownames=['True'], colnames=['Predicted'], margins=True)
```

```python
#ROC-AUC
from sklearn.metrics import roc_curve
predictions_proba = Lr.predict_proba(X_test)[:,1]
fpr, tpr, thresholds = roc_curve(y_test, predictions_proba)

plt.plot([0,1],[0,1],'go--')
plt.plot(fpr,tpr, label='Logistic Regression')
plt.xlabel('fpr')
plt.ylabel('tpr')
plt.title('Logistic Regression, ROC curve')
plt.show()
```

```python
#Classification report
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions))
```
