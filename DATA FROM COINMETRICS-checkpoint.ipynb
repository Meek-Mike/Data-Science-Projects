### CRYPTO STABLE COIN ANALYSIS

```python
#Importing Coinmetrics modul and API
from coinmetrics.api_client import CoinMetricsClient
client = CoinMetricsClient()
```

```python
#Displaying all available assets in Coinmetrics
catalog_assets_df = client.catalog_assets().to_dataframe()
only_assets_without_metrics = catalog_assets_df.dropna(subset=['metrics'])
print(only_assets_without_metrics)
```

```python
#Finding Tether networks on Coinmetrics
print(only_assets_without_metrics[only_assets_without_metrics['full_name'].str.contains('Teth')])
```

```python
#Calling out Coinmetric Asset Data using API
asset_metrics = client.get_asset_metrics(assets=['usdc','usdt_omni','usdt_eth','usdt_trx','busd'],
                                       metrics = [
                                           'PriceUSD',
                                           'AdrActCnt',
                                           'CapMVRVCur',
                                           'CapMrktCurUSD',
                                           'NVTAdj',
                                           'VelCur1yr',
                                           'TxCnt',
                                           'TxTfrValAdjUSD',
                                           'SplyCur',
                                           'HashRate'
                                       ],
                                       start_time='2023-04-03',
                                       end_time='2023-04-11',
                                      frequency ='1d'
                                       );
#Importing pandas to store Asset metrics in DataFrame
import pandas as pd
asset_metrics = pd.DataFrame(asset_metrics)
asset_metrics.head()
```

```python
Original_Data = asset_metrics
```

```python
# Information about data types for each column
print(asset_metrics.info())
```

#### Re-assigning columns with their data types

```python
# Assets
asset_metrics['asset'] = asset_metrics['asset'].astype('str')
# Time
asset_metrics['time'] = pd.to_datetime(asset_metrics['time'])
# Active AddressCount - 
asset_metrics['AdrActCnt'] = asset_metrics['AdrActCnt'].astype('int') 
# Market Cap / Realized Market Cap - The ratio of the sum USD value of the current supply to the sum "realized" USD value of the current supply.
asset_metrics['CapMVRVCur'] = asset_metrics['CapMVRVCur'].astype('float')
asset_metrics['CapMVRVCur'] = asset_metrics['CapMVRVCur'].round(decimals = 4)
# Market Cap USD - The sum USD value of the current supply. Also referred to as network value or market capitalization.
asset_metrics['CapMrktCurUSD'] = asset_metrics['CapMrktCurUSD'].astype('float')
asset_metrics['CapMrktCurUSD'] = asset_metrics['CapMrktCurUSD'].round(decimals = 3)
# NVT - he ratio of the network value (or market capitalization, current supply) divided by the adjusted transfer value
asset_metrics['NVTAdj'] = asset_metrics['NVTAdj'].astype('float')
asset_metrics['NVTAdj'] = asset_metrics['NVTAdj'].round(decimals = 4)
# Closing Price USD - The price of the asset denominated in U.S. Dollars
asset_metrics['PriceUSD'] =asset_metrics['PriceUSD'].astype('float')
asset_metrics['PriceUSD'] = asset_metrics['PriceUSD'].round(decimals = 5)
# Current Supply (Asset) - The sum of all native units ever created and currently visible on the ledger (i.e., issued) as of that day
asset_metrics['SplyCur'] = asset_metrics['SplyCur'].astype('float')
asset_metrics['SplyCur'] = asset_metrics['SplyCur'].round(decimals = 2)
#Transaction Count - The sum count of transactions that day
asset_metrics['TxCnt'] = asset_metrics['TxCnt'].astype('int')
# transfer Value, Adj (USD) - The USD value of the sum of native units transferred that interval removing noise and certain artifacts
asset_metrics['TxTfrValAdjUSD'] = asset_metrics['TxTfrValAdjUSD'].astype('float')
asset_metrics['TxTfrValAdjUSD'] = asset_metrics['TxTfrValAdjUSD'].round(decimals = 3)
# 1 Year Current Supply Velocity - the number of times that an average native unit has been transferred in the past 1 year
asset_metrics['VelCur1yr'] =asset_metrics['VelCur1yr'].astype('float')
asset_metrics['VelCur1yr'] =asset_metrics['VelCur1yr'].round(decimals = 3)
```
